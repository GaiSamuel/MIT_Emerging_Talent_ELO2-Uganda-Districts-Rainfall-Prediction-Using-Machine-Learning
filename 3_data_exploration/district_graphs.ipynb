```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# load dataset
data = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/MIT_Emerging_Talent_ELO2-Uganda-Districts-Rainfall-Prediction-Using-Machine-Learning/1_datasets/clean_data/uganda_rainfall_forecasts_2025_2026.csv"
)

data.head()  # Display the first few rows of the dataset
```

```python
# setup plot aesthetics
sns.set(style="whitegrid")
```

```python
# set the folder to save plots
output_folder = "district_plots"
os.makedirs(output_folder, exist_ok=True)
```

Generate Plots for every district and save to the graphs folder

```python
# loop through each district
for district in data['district'].unique():
    district_data = data[data['district'] == district]

    # Initialize the plot
    plt.figure(figsize=(14, 6))

    # Plot using seaborn lineplot
    sns.lineplot(
      data=district_data,
      x='date',
      y='predicted_rainfall_mm',
      marker='o',
      linewidth=2.5,
    )
    
    # Customize the plot
    plt.title(f'Predicted Rainfall Over Time in {district} 2025-2026', fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Predicted Rainfall (mm)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # replace spaces in district name with underscores for filename and save the plot to the output folder
    filename = district.replace(" ", "_")
    plt.savefig(f"{output_folder}/{filename}_predicted_rainfall.png")
    
    plt.close()  # Close the plot to free memory
    print(f"Plot saved for district: {district}")    
```
