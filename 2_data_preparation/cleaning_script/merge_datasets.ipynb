```python
import pandas as pd
import numpy as np
```

```python
dewpoint = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/dewpoint_temperature/uganda_monthly_dewpoint_temperature_1981_2025.csv"
)

precipitation = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/precipitation/uganda_monthly_precipitation_1981_2025.csv"
)

pressure = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/pressure/uganda_monthly_pressure_1981_2025.csv"
)

temperature = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/temperature/uganda_monthly_temperature_1981_2025.csv"
)

total_cloud_cover = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/totalcloudcover/uganda_monthly_total_cloud_cover_1981_2025.csv"
)

wind_u_component = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/wind_east_west/uganda_monthly_wind_u_component_1981_2025.csv"
)

wind_v_component = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/wind_north_south/uganda_monthly_wind_v_component_1981_2025.csv"
)

chirps_rainfall = pd.read_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/rainfall/uganda_districts_monthly_rainfall.csv"
)
```

```python
chirps_rainfall["date"] = pd.to_datetime(chirps_rainfall["date"])
chirps_rainfall["date"] = chirps_rainfall["date"].dt.to_period("M").astype(str)

# replace the negative values with NaN
chirps_rainfall.loc[chirps_rainfall["mean_rainfall"] < 0, "mean_rainfall"] = np.nan

chirps_rainfall
```

```python
# clean dewpoint data
dewpoint.columns = dewpoint.columns.str.lower()

dewpoint["date"] = pd.to_datetime(dewpoint["date"])
dewpoint["date"] = dewpoint["date"].dt.to_period("M").astype(str)

dewpoint["dewpoint_temperature"] = dewpoint["dewpoint_temperature"] - 273.15

# keep only dates from 1981_01 to 2025_10
dewpoint = dewpoint[(dewpoint["date"] >= "1981-01") & (dewpoint["date"] <= "2025-10")]

dewpoint = dewpoint.sort_values(by=["date", "district"], ascending=[True, True])

dewpoint.reset_index(drop=True, inplace=True)
dewpoint
```

```python
temperature.columns = temperature.columns.str.lower()
temperature["date"] = pd.to_datetime(temperature["date"])
temperature["date"] = temperature["date"].dt.to_period("M").astype(str)
temperature["temperature"] = temperature["temperature"] - 273.15
temperature = temperature[
    (temperature["date"] >= "1981-01") & (temperature["date"] <= "2025-10")
]
temperature = temperature.sort_values(by=["date", "district"], ascending=[True, True])
temperature.reset_index(drop=True, inplace=True)
temperature
```

```python
pressure.columns = pressure.columns.str.lower()
pressure["date"] = pd.to_datetime(pressure["date"])
pressure["date"] = pressure["date"].dt.to_period("M").astype(str)
pressure = pressure[(pressure["date"] >= "1981-01") & (pressure["date"] <= "2025-10")]
pressure = pressure.sort_values(by=["date", "district"], ascending=[True, True])
pressure.reset_index(drop=True, inplace=True)
pressure
```

```python
precipitation.columns = precipitation.columns.str.lower()
precipitation["date"] = pd.to_datetime(precipitation["date"])
precipitation["date"] = precipitation["date"].dt.to_period("M").astype(str)
precipitation = precipitation[
    (precipitation["date"] >= "1981-01") & (precipitation["date"] <= "2025-10")
]
precipitation = precipitation.sort_values(
    by=["date", "district"], ascending=[True, True]
)
precipitation.reset_index(drop=True, inplace=True)
precipitation
```

```python
wind_u_component.columns = wind_u_component.columns.str.lower()

wind_u_component["date"] = pd.to_datetime(wind_u_component["date"])

wind_u_component["date"] = wind_u_component["date"].dt.to_period("M").astype(str)

wind_u_component = wind_u_component[
    (wind_u_component["date"] >= "1981-01") & (wind_u_component["date"] <= "2025-10")
]

wind_u_component = wind_u_component.sort_values(
    by=["date", "district"], ascending=[True, True]
)

wind_u_component.reset_index(drop=True, inplace=True)

wind_u_component
```

```python
wind_v_component.columns = wind_v_component.columns.str.lower()

wind_v_component["date"] = pd.to_datetime(wind_v_component["date"])

wind_v_component["date"] = wind_v_component["date"].dt.to_period("M").astype(str)

wind_v_component = wind_v_component[
    (wind_v_component["date"] >= "1981-01") & (wind_v_component["date"] <= "2025-10")
]

wind_v_component = wind_v_component.sort_values(
    by=["date", "district"], ascending=[True, True]
)

wind_v_component.reset_index(drop=True, inplace=True)

wind_v_component
```

```python
# Merge all datasets on 'district', 'region', and 'date'
merged_data = (
    chirps_rainfall.merge(dewpoint, on=["district", "region", "date"], how="outer")
    .merge(temperature, on=["district", "region", "date"], how="outer")
    .merge(precipitation, on=["district", "region", "date"], how="outer")
    .merge(pressure, on=["district", "region", "date"], how="outer")
    .merge(total_cloud_cover, on=["district", "region", "date"], how="outer")
    .merge(wind_u_component, on=["district", "region", "date"], how="outer")
    .merge(wind_v_component, on=["district", "region", "date"], how="outer")
)

# Display the merged dataset
merged_data
```

```python
merged_data["date"] = pd.to_datetime(merged_data["date"])
merged_data["month"] = merged_data["date"].dt.month

valid_districts = merged_data.groupby("district")["temperature"].apply(
    lambda x: x.notna().any()
)

districts_to_keep = valid_districts[valid_districts].index

merged_data = merged_data[merged_data["district"].isin(districts_to_keep)].copy()

merged_data.reset_index(drop=True, inplace=True)

merged_data["temperature"] = merged_data.groupby(["district", "month"])[
    "temperature"
].transform(lambda x: x.fillna(x.mean()))
merged_data["surface_pressure_pa"] = merged_data.groupby(["district", "month"])[
    "surface_pressure_pa"
].transform(lambda x: x.fillna(x.mean()))

merged_data = merged_data.sort_values(by=["district", "date"]).reset_index(drop=True)

merged_data["mean_rainfall"] = merged_data.groupby("district")[
    "mean_rainfall"
].transform(lambda x: x.interpolate(method="linear", limit_direction="both"))

merged_data = merged_data.drop(columns=["month"])

merged_data["date"] = merged_data["date"].dt.to_period("M").astype(str)

merged_data
```

```python
valid_rows = merged_data.dropna(subset=["mean_rainfall", "precipitation_mm"])
conversion_ratio = (
    valid_rows["mean_rainfall"].mean() / valid_rows["precipitation_mm"].mean()
)

print(f"Conversion Ratio: {conversion_ratio:.2f}")

missing_mask = (
    merged_data["mean_rainfall"].isna() & merged_data["precipitation_mm"].notna()
)

merged_data.loc[missing_mask, "mean_rainfall"] = (
    merged_data.loc[missing_mask, "precipitation_mm"] * conversion_ratio
)

merged_data["mean_rainfall"] = merged_data.groupby("district")[
    "mean_rainfall"
].transform(lambda x: x.interpolate(method="linear", limit_direction="both"))

merged_data = merged_data.drop(columns=["precipitation_mm"])

merged_data = merged_data.rename(columns={"mean_rainfall": "rainfall_mm"})
print("\nMissing values after fixing:")
print(merged_data.isnull().sum())

merged_data
```

```python
# Save the final merged dataset to a CSV file
merged_data.to_csv(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/uganda_monthly_rainfall_and_climate_variables_1981_2025.csv",
    index=False,
)
```
