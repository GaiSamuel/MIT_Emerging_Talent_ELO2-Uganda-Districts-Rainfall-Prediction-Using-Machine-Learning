```python
files = glob.glob(
    "C:/Users/Administrator/OneDrive/Desktop/rainfall_prediction/era5/wind_east_west/*.grib"
)

datasets = []

for file in files:
    ds = xr.open_dataset(file, engine="cfgrib")
    minx, miny, maxx, maxy = ug.total_bounds
    ds_ug = ds.sel(longitude=slice(minx, maxx), latitude=slice(maxy, miny))
    ds_monthly = ds_ug.resample(time="1M").mean()
    datasets.append(ds_monthly)

combined_ds = xr.concat(datasets, dim="time")

combined_ds
```

```python
df_wind_u = combined_ds.to_dataframe().reset_index()
df_wind_u
```

```python
# convert grid cells points to shapely points
gdf = gpd.GeoDataFrame(
    df_wind_u,
    geometry=gpd.points_from_xy(df_wind_u.longitude, df_wind_u.latitude),
    crs="EPSG:4326",
)
```

```python
gdf = gpd.sjoin(gdf, ug, how="left", predicate="within")
gdf
```

```python
df_monthly = gdf.groupby(["ADM2_EN", "ADM1_EN", "time"])["u10"].mean().reset_index()
df_monthly
```

```python
# rename columns
df_monthly = df_monthly.rename(
    columns={
        "ADM2_EN": "District",
        "ADM1_EN": "Region",
        "time": "Date",
        "u10": "Wind_U_Component",
    }
)
df_monthly
```

```python
# save to csv
df_monthly.to_csv("uganda_monthly_wind_u_component_1981_2025.csv", index=False)
```
