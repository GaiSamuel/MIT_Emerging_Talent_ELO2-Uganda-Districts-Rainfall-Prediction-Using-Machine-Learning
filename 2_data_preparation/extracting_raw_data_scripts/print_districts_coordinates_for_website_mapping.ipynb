```python
import geopandas as gpd
```

```python
districts = gpd.read_file("simplified_uganda_districts.shp")
districts = districts.to_crs(epsg=3857)
districts
```

```python
# Step 2. Compute centroid in projected CRS
districts["centroid"] = districts.geometry.centroid

# Step 3. Convert back to WGS84
districts = districts.to_crs(epsg=4326)

# Step 4. Extract lat lon
districts["lat"] = districts["centroid"].to_crs(epsg=4326).y
districts["lon"] = districts["centroid"].to_crs(epsg=4326).x

# Step 5. Print JS dictionary
print("const districtLocations = {")
for idx, row in districts.iterrows():
    name = row["ADM2_EN"].lower().strip()
    lat = round(row["lat"], 4)
    lon = round(row["lon"], 4)
    print(f'    "{name}": [{lat}, {lon}],')
print("};")
```
